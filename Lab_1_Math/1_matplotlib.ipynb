> якщо робити графік по точках то кількість елментів має бути однаковою!

```python
s = 100
rng = np.random.default_rng(seed=101)
x = rng.integers(0, 5, size=s, endpoint=True)
y = rng.integers(0, 100, size=s, endpoint=True)

print(f"""
X має {x.size} елементів, з них унікальні {np.unique(x)}
По Х: Максимум {x.max()}, Мінімум {x.min()}, Середнє значення: {x.mean()}


Y має {y.size} елементів, з них унікальних {np.unique(y).size} (не будемо виводити всі елементи)
По Y: Максимум {y.max()}, Мінімум {y.min()}, Середнє значення: {y.mean()}
""")
#Homework
size = 50
rng = np.random.default_rng(seed=42)
a = rng.normal(loc=10, scale=5, size=size)
b = rng.uniform(low=0, high=50, size=size)

print(f"""
Масив A містить {a.size} елементів, унікальних значень: {np.unique(a).size}
По A: Максимум {a.max():.2f}, Мінімум {a.min():.2f}, Середнє значення: {a.mean():.2f}

Масив B містить {b.size} елементів, унікальних значень: {np.unique(b).size}
По B: Максимум {b.max():.2f}, Мінімум {b.min():.2f}, Середнє значення: {b.mean():.2f}
""")
```

> можна вивести декілька графіків та додати підписи на осі

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Випадкові величини х та у')
plt.plot(y.mean().repeat(x.max()+1), "--")

plt.show()
#Homework
plt.scatter(a, b)
plt.xlabel('A-axis')
plt.ylabel('B-axis')
plt.title('Графік випадкових A та B')
plt.plot(a.mean().repeat(b.max() + 1), "--")

plt.show()
```

> за допомогою `np.histogram` можна знайти статистичний розподіл згенерованих значень

```python
print(np.histogram(y))
plt.hist(y)
plt.show()
#Homework
print(np.histogram(a))
plt.hist(a)
plt.show()
```

> візуалізація декількох графіків

```python
x = np.linspace(0, 5, 100)
ya = (5 - x)/4
yb = (3 - x)**2
yc = (2 - x)
yd = np.full_like(x, 1)

plt.plot(x, ya, label='y = (5 - x)/4')
plt.plot(x, yb, label='y = (3 - x)**2')
plt.plot(x, yc, label='y = 2 - x')
plt.plot(x, yd, label='y = 1')

plt.title('Візуалізація')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
#Homework
a = np.linspace(-2, 4, 100)
ya = (4 - a) / 3
yb = (2 - a) ** 2
yc = (1 - a)
yd = np.full_like(a, -1)

plt.plot(a, ya, label='y = (4 - a)/3')
plt.plot(a, yb, label='y = (2 - a)^2')
plt.plot(a, yc, label='y = 1 - a')
plt.plot(a, yd, label='y = -1')

plt.title('Графіки функцій')
plt.xlabel('a')
plt.ylabel('y')

plt.legend()
plt.show()
```
